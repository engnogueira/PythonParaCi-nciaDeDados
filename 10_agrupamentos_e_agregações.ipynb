```python
import pandas as pd
```

```python
caminho = 'C:/Users/Jessica/Desktop/Arquivos_de_Exercicio/dados/mtcars.csv'

carros = pd.read_csv(caminho)
carros.head()
```

#### Renomeando colunas

```python
carros.columns = ['nomes','mpg','cyl','disp', 'hpt', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
carros.head()
```

### Grouping data by column index

#### nome_do_objeto.groupby('nome_da_serie')

♔┈♔┈♔┈( O QUE ISSO FAZ )┈♔┈♔┈♔
Para agrupar os dados de um DataFrame considerando uma coluna em particular, usamos o método `.groupby()` e passamos o nome da coluna que queremos executar o agrupamento.

```python
coluna_agrupamento = carros['cyl']
grupos_carros = carros.groupby(coluna_agrupamento)
grupos_carros.mean()
```
